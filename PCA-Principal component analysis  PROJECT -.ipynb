```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_x=scaler.fit_transform(pixels)
```

```python
scaled_x
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
principal_components=pca.fit_transform(scaled_x)
```

```python
principal_components
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8,6),dpi=100)
plt.scatter(principal_components[:,0],principal_components[:,1],c=df['number_label'])
plt.xlabel('principal_comp1')
plt.ylabel('principal_comp2')
plt.legend()
```

```python
pca3=PCA(n_components=3)
```

```python
pca3=pca3.fit_transform(scaled_x)
```

```python

fig = plt.figure(figsize=(12,8),dpi=150) # create new figure
ax = fig.add_subplot(projection='3d') # add 3d plot

# make and show plot
ax.scatter(pca3[:,0],pca3[:,1],pca3[:,2],c=df['number_label'])
plt.show()
```
