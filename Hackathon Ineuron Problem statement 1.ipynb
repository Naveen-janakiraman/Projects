```python
# feature engineering
```

```python
dataset.head()
```

```python
dataset=dataset[dataset.columns].replace(' ?',np.nan)
```

```python
dataset.columns
```

```python
dataset['country'].isna().sum()
```

```python
dataset['country'].fillna(dataset['country'].mode())
```

```python
dataset.isna().sum()
```

```python
dataset['country'].mode()[0]
```

```python
for feature in ['workclass','occupation','country']:
    dataset[feature].fillna(dataset[feature].mode()[0],inplace=True)
```

```python
dataset.isna().sum()
```

```python
# Dummies
```

```python
# ML model
```

```python
X=pd.get_dummies(dataset.drop('salary',axis=1),drop_first=True)
Y=dataset['salary']
```

```python
X.corr()
```

```python
sns.heatmap(X.corr())
```

# Decision Trees

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
scaler=StandardScaler()
```

```python
scaled_X_train=scaler.fit_transform(X_train)
scaled_X_test=scaler.transform(X_test)
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(scaled_X_train,y_train)
```

```python
y_pred=model.predict(scaled_X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_)
```

```python
# from sklearn.tree import plot_tree
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'criterion' : ["gini", "entropy"], 
              'max_leaf_nodes':[2,3,4,5],
               'max_depth':[2,3,4,5]}
```

```python
# grid_model=model()
```

```python
grid_model=GridSearchCV(estimator=model,param_grid=param_grid,cv=5)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
Y_pred=grid_model.predict(X_test)

```

```python
print(classification_report(y_test,Y_pred))
```

```python
#random forest tree
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=10,max_features='auto',random_state=101)
```

```python
model.fit(X_train,y_train)
```

```python
preds=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix,accuracy_score
```

```python
accuracy_score(preds,y_test)
```

```python
print(classification_report(preds,y_test))
```

```python
plot_confusion_matrix(model,X_test,y_test)
```

```python
test_error = []

for n in range(1,40):
    # Use n random trees
    model = RandomForestClassifier(n_estimators=n,max_features='auto')
    model.fit(X_train,y_train)
    test_preds = model.predict(X_test)
    test_error.append(1-accuracy_score(test_preds,y_test))
```

```python
plt.plot(range(1,40),test_error,label='Test Error')
plt.legend()
```

```python
n_estimators=[64,100,128,200]
max_features= [2,3,4]
```

```python
param_grid = {'n_estimators':n_estimators,
             'max_features':max_features}
```

```python
rfc=RandomForestClassifier()
grid=GridSearchCV(rfc,param_grid)
```

```python
grid.fit(X_train,y_train)
```

```python
preds=grid.predict(X_test)
```

```python
print(classification_report(preds,y_test))
```

```python
plot_confusion_matrix(grid,X_test,y_test)
```
