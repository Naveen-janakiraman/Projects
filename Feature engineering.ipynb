```python
dataset=pd.read_csv('Desktop/PYTN/train.csv')
```

```python
dataset.head()
```

```python
data_catnan=[features for features in dataset if dataset[features].isnull().sum()>1 and dataset[features].dtypes == 'O']
data_catnan
```

```python
def replace_nan(dataset,data_catnan):
    for features in data_catnan:
        dataset=dataset.copy()
        dataset[data_catnan]=dataset[data_catnan].fillna('Missing')
        return dataset
dataset=replace_nan (dataset,data_catnan)  
```

```python
dataset.head()
```

```python
data_numnan=[features for features in dataset if dataset[features].isnull().sum()>1 and dataset[features].dtypes!='O']

data_numnan
```

```python
for features in data_numnan:
    dataset[features+'nan']=np.where(dataset[features].isnull(),1,0)
    dataset[features]=dataset[features].fillna(dataset[features].median())
```

```python
dataset.head(25)
```

```python
for features in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[features]=dataset['YrSold']-dataset[features]
   
```

```python
 dataset.head()
```

```python
Num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for features in Num_features:
    dataset[features]=np.log(dataset[features])
```

```python
dataset.head()
```

## cat features

```python
cat_features=[features for features in dataset if dataset[features].dtype=='O']
cat_features
```

```python
for features in cat_features:
    temp=dataset.groupby(features)['SalePrice']
    temp=dataset.groupby(features)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.1].index
    dataset[features]=np.where(dataset[features].isin(temp_df),dataset[features],'RareVar')
```
