```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_df=scaler.fit_transform(df)
```

```python
scaled_df
```

```python
from sklearn.cluster import DBSCAN
```

```python
outliers_percent=[]
no_of_outliers=[]

for eps in np.linspace(0.001,3,50):
    
    dbscan=DBSCAN(eps=eps)
    dbscan.fit(scaled_df)
    
    no_of_outliers.append(np.sum(dbscan.labels_==-1))
    
    outliers_percent.append(100*np.sum(dbscan.labels_==-1)/len(dbscan.labels_))

fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,6))

axes[0].plot(np.linspace(0.001,3,50),no_of_outliers)
axes[0].set_title('no_of_outliers')
plt.xlim(0,3)

axes[1].plot(np.linspace(0.001,3,50),outliers_percent)
axes[1].set_title('outliers_percent')
plt.xlim(0,3)

fig.tight_layout()
```

```python
sns.lineplot(x=np.linspace(0.001,3,50),y=outliers_percent)
plt.xlim(0,3)
plt.ylim(0,40)
plt.hlines(xmin=0,xmax=3,y=3,colors='red',ls='--')
plt.ylabel("Percentage of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
```

```python
# so we can choose eps =2
```

```python
dbscan=DBSCAN(eps=2)
```

```python
labels=dbscan.fit_predict(scaled_df)
```

```python
df['labels']=labels
```
