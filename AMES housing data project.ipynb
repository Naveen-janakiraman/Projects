```python
ames=pd.read_csv('C:\\documents\\study material\\data science\\Machine learning udemy notes\\DATA/AMES_Final_DF.csv')

ames.head()
```

```python
ames.info()
```

```python
X=ames.drop('SalePrice',axis=1)
y=ames['SalePrice']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNetCV
```

```python
model=ElasticNet()
```

```python

param_grid = {'alpha':[0.1,1,5,10,50,100],
              'l1_ratio':[.1, .5, .7, .9, .95, .99, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model=GridSearchCV(estimator=model,param_grid=param_grid,cv=5,scoring='neg_mean_squared_error',verbose=2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error

Y_pred=grid_model.predict(X_test)

MAE=mean_absolute_error(y_test,Y_pred)
RSME=np.sqrt(mean_squared_error(y_test,Y_pred))
```
