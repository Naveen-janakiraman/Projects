```python
def waiting_mean(arrival_time,service_time):
    waiting_time=[]
    leaving_time=[]
    
    waiting_time.append(0)
    leaving_time.append(arrival_time[0]+service_time[0]+waiting_time[0])
    
    for i in range(1,len(arrival_time)):
        waiting_time.append(max(0,(leaving_time[i-1]-arrival_time[i])))
        leaving_time.append(arrival_time[i]+service_time[i]+waiting_time[i])
        
    mean_waiting=np.mean(waiting_time)
    return mean_waiting
```

```python
# Factory - supply and demand
# minimize shipping cost
```

```python
model=LpProblem('shipping',LpMinimize)
```

```python
customers= ['aus','ind','swe']
factory=['fac1','fac2']
products=['chair','table','beds']
```

```python
keys=[(f,p,c) for f in factory for p in products for c in customers]
```

```python
keys
```

```python
var = LpVariable.dicts('shipping',keys,0,None,cat='Integer')
```

```python
var
```

```python
cost_values= [50,80,50,
             60,90,60,
             70,90,70,
              80,50,80,
             90,60,90,
             90,70,90]
```

```python
costs=dict(zip(keys,cost_values))
```

```python
costs
```

```python
demand_keys=[(p,c) for c in customers for p in products ]
```

```python
demand_values=[50,80,200,120,80,40,30,60,175]
```

```python
demand=dict(zip(demand_keys,demand_values))
```

```python
# objective function
```

```python
model+= lpSum(var[(f,p,c)]*costs[(f,p,c)] for f in factory for p in products for c in customers)
```

```python
#constriants
```

```python
model+= lpSum(var[('fac1',p,c)]for p in products for c in customers) <=500
```

```python
model+= lpSum(var[('fac2',p,c)]for p in products for c in customers) <=500
```

```python
for c in customers:
    for p in products:
        model+= (var[('fac1',p,c)]+var[('fac2',p,c)]) >= demand[(p,c)]
```

```python
model.solve()
```

```python
value(model.objective)
```

```python
list1=[]
for i in range(1,13):
    list1.append(i)
```

```python
param['t']=range(1,13)
```

```python
param['t']=list1
```

```python
param
```

```python
param=param.set_index('t')
```

```python
param=param.rename(columns={'Unnamed: 0':'Period'})
```

```python
param
```

```python
list2=[]
for i in range(0,13):
    list2.append(int(i))
```

```python
inventory=LpVariable.dicts('inv',list2,0,None,'Integer')
inventory[0]=200
```

```python
production=LpVariable.dicts('prod',list1,0,None,'Integer')
binary=LpVariable.dicts('bin',list1,0,None,'Binary')
```

```python
time=list1
```

```python
model=LpProblem('prod',LpMinimize)
```

```python
# objective
```

```python
model+= lpSum([inventory[t]*param.loc[t,'storage cost']+production[t]*param.loc[t,'var']+binary[t]*param.loc[t,'fixed cost'] for t in time ])
```

```python
for t in time:
    model+= production[t]-inventory[t]+inventory[t-1] >= param.loc[t,'demand']
    model+= production[t] <= binary[t]*param.loc[t,'Capacity']
```

```python
model.solve()
```

```python
value(model.objective)
```

```python
for i in production:
    print(production[i],production[i].varValue )
```

```python
for v in model.variables():
    print(v,v.varValue)
```
